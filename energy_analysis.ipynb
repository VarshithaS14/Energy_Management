```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Load the dataset
df = pd.read_csv("household_energy.csv",parse_dates=['timestamp'])
df
```

```python
#Extravt time features
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['weekday'] = df['timestamp'].dt.dayofweek
df['date'] = df['timestamp'].dt.date
df
```

```python
#Basic summary
print(" === Basic Info ===")
print(df.info())
```

```python
print("\n=== Summary ===")
print(df.describe())
```

```python
#Daily and Weekly Consumption Analysis
daily_consumption = df.groupby('date')['energy_consumption'].sum()
weekly_consumption = df.groupby(df['timestamp'].dt.isocalendar().week)['energy_consumption'].sum()
```

```python
print("\n=== Daily Consumption (Frist 5 Days) ===")
print(daily_consumption.head())
```

```python
#visualize Daily usage
plt.figure(figsize=(12, 4))
plt.plot(daily_consumption.index, daily_consumption.values, marker=0)
plt.title("Daily Household Energy Consumption")
plt.xlabel("Date")
plt.ylabel("Energy (kWh)")
plt.grid(True)
plt.tight_layout()
plt.savefig("daily_energy_plot.png")
plt.show()
```

```python
print("\n=== Weekly Consumption  ===")
print(weekly_consumption.head())
```

```python
#visualize weekly usage
plt.figure(figsize=(12, 4))
plt.plot(weekly_consumption.index, weekly_consumption.values, marker='o', linestyle='-')
plt.title("weekly Household Energy Consumption")
plt.xlabel("week")
plt.ylabel("Energy (kWh)")
plt.grid(True)
plt.tight_layout()
plt.savefig("weekly_energy_plot.png")
plt.show()
```

```python
#Forecasting with Linear Regression
features = ['temperature', 'outside_temperature', 'device_usage','hour', 'weekday']
x = df[features]
y = df['energy_consumption']
```

```python
#ML model
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
joblib.dump(model,"forecast_model.pkl")
```

```python
y_pred=model.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_test, y_pred))
print(f"\nForecating RMSE: {rmse:.3f}")
```

```python
df['predicted_consumption']=model.predict(x)
df
```

```python
df['moving_avg'] = df['energy_consumption'].rolling(window=6,min_periods=1).mean()
df['inefficient'] = df['energy_consumption']>1.25 * df['moving_avg']
```

```python
print("\n=== Inefficient Usage Entries ===")
ineff_df = df[df['inefficient']]
print(ineff_df[['timestamp', 'energy_consumption', 'moving_avg']].head())
```

```python
# Load your data
ineff_df = pd.read_csv("household_energy.csv")

# Extract hour from timestamp
ineff_df['hour'] = pd.to_datetime(ineff_df['timestamp']).dt.hour
```

```python
# Recommendation Engine
recommendations = []

for _, row in ineff_df.iterrows():
    if row['device_usage'] > 0 and row['energy_consumption'] > 3:
        recommendations.append("Consider turning off idle devices during high usage hours.")

    if row['outside_temperature'] > 30 and row['energy_consumption'] > 3:
        recommendations.append("High outside temp + high usage: Adjust thermostat settings.")

    if row['hour'] in [0, 1, 2, 3, 4] and row['energy_consumption'] > 2:
        recommendations.append("High energy use during late night hours: Reduce appliance usage.")
```

```python
recommendations = list(set(recommendations))
print("\n=== Recommendations ===")
for rec in recommendations:
    print(f"- {rec}")
```
